```python
import pandas_datareader as pdr
import pandas as pd 
from datetime import datetime
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
data = yf.download('TSLA', period='5y')
print(data)
```

data['Open'].plot(figsize=(15,6))

```python
data = data.sort_index()
```

```python
data.head()
```

```python
data['Open:10 days rolling'] = data['Open'].rolling(window=10, min_periods=1).mean()
```

```python
data
```

```python
data[['Open','Open:10 days rolling']].plot(xlim=['2021-05-10','2022-05-10'],figsize= (15,6))
```

```python
data['Open:30 days rolling'] = data['Open'].rolling(window=30, min_periods=1).mean()
```

```python
data
```

```python
data['Open:50 days rolling'] = data['Open'].rolling(window=50, min_periods=1).mean()
```

```python
data
```

```python
data[['Open','Open:10 days rolling','Open:30 days rolling','Open:50 days rolling']].plot(xlim=['2021-05-10','2022-05-10'],figsize =(15,6))
```

```python
##Expanding
###  CMA
data['Open'].expanding().mean().plot(figsize=(15,5))
```

```python
#EMA  Exponential Moving Average
# let smoothing factor -0.1
data['EMA_0.1'] = data['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
data[['Open','EMA_0.1']].plot(xlim=['2021-05-10','2022-05-10'],figsize =(15,6))
```

```python
#EWMA  Exponential weighted Moving Average
# let smoothing factor -0.1
data['EMA_5days'] = data['Open'].ewm(span=5).mean()
```

```python
 data[['Open','EMA_5days','EMA_0.1']].plot(figsize=(15,5))
```
